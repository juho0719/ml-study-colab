## [이론1] 인공지능과 머신러닝이란

### 1. 인공지능이란
- **인공 + 지능**이라는 의미 그대로, 인공적으로 만들어진 지능을 가지고 일을 수행하는 시스템
- **강 인공지능(Strong AI)**: 인공적인 의식을 가지고 실체 인간과 유사하게 일을 수행하는 인공지능
- **약 인공지능(Weak AI)**: 인공적인 의식과는 별개로, 단지 유용한 도구로써 사용되는 인공지능
- 강 인공지능은 **사람과 비슷하게 행동하고 생각**
- 약 인공지능은 **특정 목적을 위해서 설계**
  - 안면인식, AI 스피커, 유튜브 추천 알고리즘 등

### 2. 머신러닝이란
- **컴퓨터가 학습할 수 있도록 하는 알고리즘과 기술을 개발하는 것**
- 머신러닝은 인공지능의 한 분야이며, 빅데이터를 분석하는 다양한 기법 중 하나
- 머신러닝은 **번역 이전 문장과 번역 문장으로 이루어진 데이터를 가지고 학습**
  - 마치 사람처럼 과거에 배웠던 문장을 활용하여 문맥에 맞는 번역
  - 번역하는 판단 근거는 학습에서 사용되었던 데이터
- 장점
  - 어떤 문제를 풀기 위해 기존엔 사람이 일일이 규칙을 정해줘야 했지만, 머신러닝은 데이터 기반의 학습을 통해 **그 규칙을 스스로 찾을 수** 있음
  - 고전적 방식으로 해결하기 힘들었던 **복잡한 문제에 대한 해결 방법**을 머신러닝 모델을 이용해 찾을 수 있음

### 3. 문제

Q. 인공지능에 대한 설명으로 옳지 않은 것을 모두 고르면?

- [ ] 인공지능은 인공적으로 만들어진 지능을 갖고 일을 수행하는 시스템이다.
- [X] 강 인공지능의 예시로 AI 스피커가 있다.
- [X] 약 인공지능은 사람과 비슷하게 행동하고 생각하는 인공지능이다.
- [ ] 약 인공지능의 예시로 안면 인식 시스템이 있다.

해설

**강 인공지능(Strong AI)**은 인공적인 의식을 가지고 실제 인간과 유사하게 일을 수행.
**약 인공지능(Weak AI)**은 인공적인 의식과는 별개로, 단지 유용한 도구로써 사용되는 인공지능. 따라서 AI 스피커는 약 인공지능의 예시



Q. 머신러닝에 대한 설명으로 옳지 않은 것은?
- [ ] 머신러닝은 컴퓨터가 데이터를 학습할 수 있도록 알고리즘과 기술을 개발하는 것을 말한다.
- [ ] 머신러닝은 인공지능의 한 분야이며, 빅데이터를 분석하는데도 사용된다.
- [X] 머신러닝을 사용한 번역 모델은 단어를 1대1로 매치하여 번역한다.
- [ ] 기존 방식으로 해결하기 힘들었던 문제를 머신러닝을 통하여 풀 수 있다.

해설


단어를 1대1로 매치하여 번역하는 것은 기존의 자동 번역기 모델의 방식. 머신러닝 번역 모델은 번역 이전 문장과 번역된 문장으로 이루어진 데이터를 학습한 후, 이를 활용하여 문맥에 맞게 문장을 번역
