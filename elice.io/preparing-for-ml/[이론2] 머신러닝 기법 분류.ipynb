## [이론2] 머신러닝 기법 분류
- 머신러닝 기법은 **지도 학습, 비지도 학습, 강화 학습**으로 나뉨

### 1. 지도 학습
- 예측하고 싶은 대상에 대한 정보를 의미하는 **feature(특성) 데이터**와 그 결과값에 해당하는 **label(레이블) 데이터**를 활용하여 학습하는 모델
- 대표적인 지도 학습 모델의 예시로, 스팸 메일을 분류하는 머신러닝을 모델을 들 수 있음
  - feature데이터는 `작성자`, `발신 날짜 시간`, `메일 내용`, label 데이터는 `스팸 여부`가 있음
  - 머신러닝 분류 보델은 이처럼 준비된 feature, label 데이터를 학습한 후 새로운 메일이 스팸인지 아닌지를 판별
  - feature 데이터에 따라 특정 label(스팸인지 아닌지)로 나누는 방식을 지도 학습 방식 중 **분류(classification)**
- 온도와 같이 **실수로 되어 있는 값**이 label 데이터로 주어져 있다면 **회귀(regression)**방식으로 해결
- 회귀 방식에 관한 예시로 면적에 따라 땅 가격을 예측하는 머신러닝 모델이 있다고 가정
  - 면적 데이터는 feature
  - 땅 가격은 label
  - feature 데이터에 해당하는 면적의 크기에 따라 예측된 땅 가격을 찾을 수 있음
- **지도 학습은 머신러닝 모델에게 label 데이터를 가르쳐주고 학습시키는 방식**

### 2. 비지도 학습
- **label 데이터 없이 feature 데이터만을 사용하여 학습**
- 대표적 기법은 **클러스터링**
  - 동네 앞 공원에 서식하는 곤충들에 대해서 몸통 크기와 날개 면적을 잰 후 데이터화했다고 가정
  - 무슨 곤충인지 모르지만 비슷한 몸통 크기와 날개 면적을 가지고 있다면 같은 종의 곤충이라 생각할 수 있음
  - 비슷한 특징을 갖는 곤충끼리의 집합인 클러스터(군집: cluster)를 만들어 묶는다면 새로 발견된 곤충도 어떤 종인지 예측할 수 있음
  - `몸통 크기`,`날개 면적`과 같은 feature데이터들을 어떠한 규칙으로 묶어내는 방법을 **클러스터링(clustering)**방식이라 함
- 만약 곤충 하나에 대해서 30종류의 feature가 존재한다면 클러스터링 하기 쉽지 않음
- **차원 축소(dimensionality reduction)**를 통하여 30가지 feature를 형태는 다를 수 있지만, 2-3가지 새로운 feature로 줄여줄 수 있음
- 또 하나 예시로 **추천 시스템(recommender system)**이 있음
- 유튜브나 넷플릭스에서 자신도 모르게 추천되는 영상이 나옴
- 영상뿐만 아니라 쇼핑몰에서도 쇼핑했던 데이터를 바탕으로 다른 제품을 추천하는 광고를 보여줌
- 추천 시스템은 사용자가 선호할 만한 아이템을 추측함으로써 사용자에게 적합한 특정 항목을 제공

### 3. 강화 학습
- **학습 데이터를 사용하지 않고 학습**
- 알파고의 업그레이드 버전인 알파고 제로처럼 바둑 전략을 모르는 어린아이가 바둑 규칙만 알고 두는 것 처럼, 스스로 바득을 두면서 데이터를 쌓아감
- 처음에는 전략 자체가 없기에 랜덤하게 두지만, 바둑이 끝나고 승패 및 점수를 보면서 스스로 내가 두었던 방법이 좋았는지 안 좋았는지를 판단
- 이러한 과정이 반복되면서 점점 더 승률이 높은 방식으로 바둑을 두게 됨
- 이처럼 강화 학습은 **에이전트가 리워드를 최대한 받을 수 있는 액션을 하도록 학습 과정**을 의미
- 강화 학습 용어
  - **에이전트**: 주어진 문제 상황에서 행동하는 주체
  - **상태**: 현재 시점에서 상황이 어떤지를 알려주는 값들의 집합
  - **액션**: 상태를 변화시킬 수 있는 에이전트의 행동들
  - **리워드**: 에이전트의 액션에 따라 얻을 수 있는 보상을 값으로 수치화한 것

### 4. 신경망 모델
- 인간의 신경망과 비슷한 형태인 **인공 뉴런을 사용한 모델**
- 신경망 모델은 **퍼셉트론(Perceptron)**이라는 인공 신경을 엮어 학습에 사용
- **더욱 복잡한 구조의 신경망 모델을 학습시키는 알고리즘을 딥러닝**이라 부름
- 딥러닝을 포함한 인공 신경망 기법은 다른 머신러닝 기법과 다르게 **데이터 특성들의 관계를 이해하지 못하여도 동작**
- 딥러닝 기술은 이미지 처리 분야 및 자연어 처리 분야에 더욱 많이 사용되기 시작
- 이미지 처리 부야는 초기에 사진의 label을 분류하는 것으로 시작해 물체를 인식하고 새로운 이미지를 생성하고 분석하는 등 다양한 영역으로 발전
- 자연어 처리는 번역 모델을 시작으로 대화가 가능한 챗봇, 문서 요약, 음석 인식 분야로 확장되어 말하는 인공지능의 기반을 다짐
- 딥러닝의 단점은 딥러닝 자체를 수행하기 위해서는 높은 코스트가 필요
- 단순하고 확실한 모델들을 사용하는 것이 더 좋은 결과를 낼 수도 있음
- 딥러닝 모델의 원리는 완전히 밝혀지지 않았기에 모델을 분석하는 면에서는 기존의 머신러닝 기법이 좋음
