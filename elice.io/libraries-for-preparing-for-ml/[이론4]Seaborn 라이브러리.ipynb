# Seaborn 라이브러리

- **seaborn**은 matplotlib기반으로 하여 다양한 색상 테마, 통계용 차트 기능이 추가된 패키지

### 0. 라이브러리 import 및 데이터 불러오기
- seaborn의 `load_dataset()` 또는 pandas의 `read_csv()`를 통해 데이터를 로드 할 수 있음
- 붓꽃 데이터인 'iris'는 seaborn 라이브러리에서 제공하는 데이터 중 하나


````python
# Seaborn 라이브러리 import 하기 
import seaborn as sns 

# iris (붓꽃) 데이터 불러오기
df=sns.load_dataset('iris') 

# 전체 데이터에서 처음 5개의 row 데이터 표시 (내용 확인)
df.head() 
````

### 1. 산점도와 분포 근사 선 그래프 그리기 - regplot()
- `regplot()`은 산점도뿐만 아니라 분포를 근사하는 ㅅ너도 함께 출력
- `regplot()` 대표 파라미터
  - **x**: x축 데이터
  - **y**: y축 데이터
  - **color**: 색상 설정
  - **line_kws**: 근사선의 특성 설정 (Python Dictionary 형태)
  - **scatter_kws**: 데이터점들의 특성 설정 (Python Dictionary 형태)
- 아래 코드에서 두 변수 sepal_length와 petal_length는 각각 꽃받침의 길이와 꽃잎의 길이를 의미

````python
import matplotlib.pyplot as plt

sns.regplot(x=df["sepal_length"], y=df["petal_length"])
plt.show()
````

- **color**로 색상 변경 가능

````python
sns.regplot(x=df["sepal_length"], y=df["petal_length"], color='red') 
plt.show()
````

- **line_kws**로 근사선의 속성만 변경 가능

````python
sns.regplot(x=df["sepal_length"], y=df["petal_length"], line_kws={'color': 'red'})
plt.show()
````

- **scatter_kws**로 점의 속성만 변경 가능

````python
sns.regplot(x=df["sepal_length"], y=df["petal_length"], scatter_kws={'color': 'green'})
plt.show()
````

- [regplot()에 대해 더 알아보기](https://seaborn.pydata.org/generated/seaborn.regplot.html?highlight=regplot#seaborn.regplot)

### 2. 빈도수 막대그래프 그리기 - countplot()
- `countplot()` 대표 파라미터
  - **x**: 열 이름(문자열)
  - **data**: 시각화를 위한 데이터(Pandas DataFrame형)
- 아래 코드는 타이타닉 탑승자 데이터를 가져와 "class"열의 빈도수를 출력하는 코드

````python
import seaborn as sns

# 타이타닉 데이터 불러오기
titanic = sns.load_dataset("titanic") 

# 전체 데이터에서 처음 5개의 row 데이터 표시 (내용 확인)
titanic.head()
````

- 파라미터 **x**에는 출력을 위한 열이름인 "class"를 입력하고, **data**에는 데이터 프레임인 titanic을 입력 

````python
ax = sns.countplot(x="class", data=titanic)
plt.show()
````

- 파라미터 **x**대신 **y**를 사용하면 y축을 기준으로 출력된 막대그래프

````python
ax = sns.countplot(y="class", data=titanic)
plt.show()
````

- [countplot()에 대해 더 알아보기](https://seaborn.pydata.org/generated/seaborn.countplot.html?highlight=countplot#seaborn.countplot)

### 3. 산점도와 히스토그램 그래프 그리기 - jointplot()
- `jointplot()` 대표 파라미터
  - **x,y**: 열과 행 이름(문자열)
  - **data**: 시각화를 위한 데이터(Pandas DataFrame형)
  - **kind**: 차트 종류 (기본은 scatter)
- 아래 코드는 레스토랑에서의 팁 가격에 대한 데이터를 불러와 `jointplot()`으로 출력

````python
sns.set(style="white") # 배경 색상 설정

# 팁 데이터 불러오기
tips = sns.load_dataset("tips")

# 전체 데이터에서 처음 5개의 row 데이터 표시 (내용 확인)
tips.head()
````

````python
g = sns.jointplot(x="total_bill", y="tip", data=tips)
plt.show()

# reg (regression + scatter) : 근사선
g = sns.jointplot("total_bill", "tip", data=tips, kind="reg") 
plt.show()
````

- [jointplot()에 대해 더 알아보기](https://seaborn.pydata.org/generated/seaborn.jointplot.html?highlight=jointplot#seaborn.jointplot)
