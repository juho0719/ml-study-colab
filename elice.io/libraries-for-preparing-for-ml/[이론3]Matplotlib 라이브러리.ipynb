
# Matplotlib 라이브러리
- **데이터를 시각화하여 데이터의 전체적인 분포나 숨겨진 특징들을 알아보는 것**
- 라이브러리 내부에는 시각화를 위한 간단한 선 그래프부터 산점도, 히스토그램 등 다양한 함수 제공

### 0. 라이브러리 import 하기
- import 할 때는 주로 `plt`라는 명칭으로 사용

````python
import matplotlib.pyplot as plt 
````

### 1. 기본 그래프 그리기 - plot()
- `plot()`함수는 입력한 데이터의 포인트들을 연결하여 직선 그래프를 출력하는 함수

````python
import matplotlib.pyplot as plt

plt.plot(["Seoul", "Paris", "Seattle"], [30, 25, 55])
plt.xlabel('City')
plt.ylabel('Response')
plt.title('Experiment Result')
plt.show()
````

- 첫 번째 입력 데이터인 ["Seoul", "Paris", "Seattle"]은 x축에 해당하는 각 포인트를 의미
- 문자열(string형)로 입력받은 경우에는 순서에 따라 x축에 입력
- 두 번째 입력 데이터인 [30,25,55]는 y축에 해당하는 각 포인트를 의미
- `plot()`함수는 각 포인트들을 연결하여 직선 그래프를 출력
- 그래프의 색상이나 포인트의 도형 형태 등도 설정할 수 있음
  - [그래프 설정하기](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html?highlight=plot#matplotlib.pyplot.plot)

- `plot()`과 관련된 함수들
  - `.xlabel()`: x축 label 명칭 설정
  - `.ylabel()`: y축 label 명칭 설정
  - `.title()`: 타이틀 명칭 설정
  - `.show()`: 그래프 출력 (jupyter notebook 환경에선 show() 함수 사용 없이도 출력 가능)
  - [그 외 다양한 함수 확인하기](https://matplotlib.org/api/pyplot_summary.html)

### 2. 산점도 그래프 그리기 - scatter()
- 연결되지 않은 포인트들을 출력

````python
import matplotlib.pyplot as plt 

plt.scatter([1, 3, 2], [0, -1, 2])
plt.show()
````

- `scatter()`함수는 `plot()`함수와 동일하게 x축 데이터와 y축 데이터를 입력받고, 색상과 사이즈 등의 파라미터를 조절할 수 있음
- [그래프 설정 방법 알아보기](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html?highlight=scatter#matplotlib.pyplot.scatter)

### 3. 히스토그램 그래프 그리기 - hist()
- 히스토그램 그래프를 출력


````python
import matplotlib.pyplot as plt

plt.hist([1, 3, 2, 2, 3, 3, 3, 1, 4, 8, 2, 6, 6], bins=None)
plt.show()
````

- `hist()`함수의 파라미터 중 **bins**는 x축의 분포를 몇 개로 나누어 출력할 지 의미

````python
plt.hist([1, 3, 2, 2, 3, 3, 3, 1, 4, 8, 2, 6, 6], bins=2)
plt.show()
````

- bins를 2로 설정하면 4.5를 기준으로 2개의 range안에서 개수를 출력
- [`hist()`에 대해 더 알아보기](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html?highlight=hist#matplotlib.pyplot.hist)

### 4. 막대 그래프 그리기 - bar()
- `bar()`함수는 각 x축 포인트에 따른 y축 값을 막대그래프를 그려주는 함수
- `hist()`함수의 경우에는 입력 데이터의 **분포**를 표현하는 데에 비해 `bar()`함수는 입력되는 각 포인트에 대해 막대로 표현
- `plot()`함수의 데이터를 `bar()`함수를 이용해서 막대그래프로 표현


````python
import matplotlib.pyplot as plt 

plt.bar(["Seoul", "Paris", "Seattle"], [30, 25, 55])
plt.xlabel('City')
plt.ylabel('Response')
plt.title('Experiment Result')
plt.show()
````

- [`bar()`에 대해서 더 알아보기](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html?highlight=bar#matplotlib.pyplot.bar)

### 5. 한 공간에 여러 개의 그래프 출력하기
- 여러 그래프를 동시에 출력하기 위해서는 아래 예시처럼 연속적으로 여러 개의 그래프 함수를 사용하고 마지막에 `.show()` 함수를 사용


````python
import matplotlib.pyplot as plt 

plt.plot([1, 2, 3], [1, 4, 9])
plt.plot([2, 3, 4], [5, 6, 7])
plt.bar([1,2,4],[2, 3, 4])
plt.xlabel('Sequence')
plt.ylabel('Time(secs)')
plt.title('Result')
plt.legend(['Elice', 'AI', 'ML'])
plt.show()
````

### 6. 여러 결과 창 묶어서 출력하기 - subplot()
- `subplot()`함수는 여러 결과 창을 묶어서 출력

````python
# 임의의 값을 생성하기 위해 numpy 라이브러리 임포트 하기
import numpy as np

x = np.random.rand(5)
y = np.random.rand(5)

plt.subplot(221)
plt.scatter(x, y, s=80, c='r', marker=">")

plt.subplot(222)
plt.scatter(x, y, s=80, c='b', marker=(5, 0))

verts = np.array([[-1, -1], [1, -1], [1, 1], [-1, -1]])
plt.subplot(223)
plt.scatter(x, y, s=80, c='k', marker=verts)

plt.subplot(224)
plt.scatter(x, y, s=80, c='y', marker=(5, 1))

plt.show()
````

- `subplot()`함수의 입력값으로 자연수 3개가 필요
- 예시에서 사용된 **subplot(221)**은 (2,2)행렬 출력창에서 1번째 출력창으로 설정하는 것을 의미
- `plt.subplot(221)`을 그래프 함수에 앞서 입력해야 해당 그래프의 위치를 조정할 수 있음
- [`subplot()`에 대해 더 알아보기](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html?highlight=subplot#matplotlib.pyplot.subplot)
