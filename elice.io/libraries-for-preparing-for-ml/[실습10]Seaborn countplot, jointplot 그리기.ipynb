# Seaborn countplot, jointplot 그리기

- 도수 분포를 나타내는 `countplot()`, 산점도와 도수 분포를 모두 나타내는 `jointplot()`함수를 사용하여 `tips`데이터를 시각화

### countplot과 jointplot을 활용하기 위한 함수/라이브러리
- `countplot('x', data)`: 데이터 명에 해당하는 각 카테고리 값 별 빈도수를 표시하는 막대 그래프를 출력
  - `x`: x축의 데이터로 열 이름 입력
  - `data`: 시각화를 위한 데이터(Pandas DataFrame형)
- `jointplot('x_axis', 'y_axis', data, kine)`: 데이터 산점도와 히스토그램을 함께 출력
  - `x_axis`: 행 이름
  - `y_axis`: 열 이름
  - `data`: 시각화를 위한 데이터(Pandas DataFrame형)
  - `kind`: 차트 종류(기본은 scatter)

---

### 실습
- `countplot()`함수의 출력물을 `sns_plot_size`로 저장("size"데이터에 해당하는 자료를 x축으로 출력)
- `jointplot()`함수의 출력을물 `g`로 저장("total_bill"을 x축, "tip"을 y축, "resid"형태 그래프로 설정)

```python
# Seaborn 라이브러리 import 하기 
import seaborn as sns 
import matplotlib.pyplot as plt 

# seaborn의 load_dataset을 사용하여 tips (팁 가격) 데이터 불러오기
df = sns.load_dataset('tips') 

# countplot함수의 출력물을 sns_plot_size으로 저장
# "size" 데이터에 해당하는 자료를 x축으로 출력
sns_plot_size = sns.countplot(x="size", data=df)

# jointplot함수의 출력물을 g로 저장
# "total_bill"을 x축, "tip"을 y축, "resid" 형태 그래프로 설정 
g = sns.jointplot(x="total_bill", y="tip", data=df, kind="resid")

# 엘리스 플랫폼 내의 출력을 위한 함수
fig = sns_plot_size.get_figure()
fig.savefig("plot_siz.png")

g.savefig("plot.png")
```
